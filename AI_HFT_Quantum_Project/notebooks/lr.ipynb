```python
import torch
import torch.nn as nn
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Load the data into a pandas DataFrame
df = pd.read_csv(filepath_or_buffer="/Users/shamweelmohammed/Desktop/Masters/Hackathons/AIHackathon24/AI_HFT_Quantum_Project/data/processed/updated_Binance USD.csv")
```

```python
def load_data(df):
    """Load and prepare data from a dictionary."""
    X = df[['Open', 'High', 'Low', 'Volume']]
    y = df['Close']
    return X, y

def train_model(X, y):
    """Train a linear regression model."""
    model = LinearRegression()
    model.fit(X, y)
    return model

def make_predictions(model, X):
    """Make predictions using the trained model."""
    return model.predict(X)

def evaluate_model(model, X, y):
    """Evaluate the model's performance."""
    predictions = model.predict(X)
    mse = mean_squared_error(y, predictions)
    return mse
```

```python
from joblib import dump, load

def save_model(model, filename):
    """Save the model to a file."""
    dump(model, filename)

def load_model(filename):
    """Load a model from a file."""
    return load(filename)

# Using the previously defined functions for training and then saving the model
X, y = load_data(df)
model = train_model(X, y)
save_model(model, 'stock_price_model.joblib')

# Load the model later for making predictions
loaded_model = load_model('stock_price_model.joblib')
predictions = make_predictions(loaded_model, X)

print("Predictions:", predictions)
```
